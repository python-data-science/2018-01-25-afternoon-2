```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
%matplotlib inline
```

```python
df = pd.read_csv("lemon_training.csv")
df.head(3)
```

```python
y = df.IsBadBuy
X = df.drop(['SubModel','MMRAcquisitionAuctionAveragePrice','MMRAcquisitionRetailAveragePrice','MMRCurrentRetailCleanPrice','MMRCurrentAuctionAveragePrice','IsOnlineSale','MMRCurrentRetailAveragePrice','PurchDate','VehYear', 'RefId', 'IsBadBuy', 'Auction', 'Trim', 'VNST', 'Color', 'WheelTypeID', 'WheelType', 'VehOdo', 'Size', 'PRIMEUNIT', 'BYRNO',  ], axis = 1)
X = pd.get_dummies(X)
X = X.fillna(0)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = GradientBoostingRegressor(n_estimators=100).fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.feature_importances_
```

```python
list(zip(X_test, model.feature_importances_))
```

```python
cv = KFold(n_splits=5)
for estimators in range(100, 1000, 100):
    for depth in range(1, 4):
        model = GradientBoostingRegressor(n_estimators=estimators, max_depth=depth)
        scores = []
        for train_i, test_i in cv.split(X):
            Xr, yr, Xt, yt = X.loc[train_i], y.loc[train_i], X.loc[test_i], y.loc[test_i]
            model.fit(Xr, yr)
            scores.append(model.score(Xt, yt))
        print('depth:', depth, 'estimators:', estimators, 'scores:', sum(scores)/len(scores), 'min:', min(scores), 'max:', max(scores))
```
